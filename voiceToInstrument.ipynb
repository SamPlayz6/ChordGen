```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

# We'll need IPython.display's Audio widget
from IPython.display import Audio
```

```python
#Read in data
y, sr = librosa.load('C:\\Users\\user\\Documents\\GitHub\\ChordGen\\SungMelodies\\M1.wav',   sr = 22050, mono = True )
```

```python
#Mel-spectrogram
# Passing through arguments to the Mel filters
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)

D = np.abs(librosa.stft(y))**2
S = librosa.feature.melspectrogram(S=D, sr=sr)

librosa.feature.melspectrogram(y=y, sr=sr)
fig, ax = plt.subplots()
s_db = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(s_db, x_axis='time', y_axis='mel', sr=sr,fmax=8000, ax=ax)

fig.colorbar(img, ax=ax, format= '%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
plt.show()
```

```python
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)

def detect_pitch(t):
  index = magnitudes[:, t].argmax()
  pitch = pitches[index, t]

  return pitch

print(len(pitches),pitches[0])


#Generate list of max pitches per unit time
max_pitch_array = []
# for i in range(len(pitches)-1):
i = 0
while True:
  i += 1
  try:
    max_pitch_array.append(detect_pitch(i))
  except:
    break
```

```python
#Playback - Original Audio
print(y)
print(len(y),sr)

Audio(data=y, rate=sr)
```

```python
# Playback - Transformed Audio
print(len(max_pitch_array),max_pitch_array)

Audio(data=max_pitch_array, rate=43)


#Still working on converting the voice to a simple pitch audio file. With Piano for example
```
